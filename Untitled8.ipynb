```python
import numpy as np
from math import factorial

y = np.array([7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 2, 
              3, 3, 1, 2, 2, 3, 1, 4, 0, 8, 3, 
              5, 3, 0, 0, 7, 1, 1, 3, 1, 3, 2, 
              4, 1, 6, 2, 2, 4, 1, 3, 1, 1, 2, 
              7, 3, 3, 2, 2, 2]) 
M=10
eps = 0.00005
iter = 0

x = np.random.rand()
pi0 = np.array([x,1-x])
mu0 = np.random.rand(2)
pi1 = np.array([1,1])
mu1 = np.array([1,1])


while np.abs(pi1-pi0).sum() + np.abs(mu1-mu0).sum() > eps:
    print('iteration = ' + str(iter), 'pi = ' + str(pi0), 'mu = ' + str(mu0), sep='\n', end='\n\n')
    pi1 = pi0
    mu1 = mu0
    iter += 1
    k = []
    for i in y:
        l = []
        for j in mu0:
            l.append((factorial(M) // (factorial(i) * factorial(M-i))) * j ** i * (1-j) ** (M-i))
        k.append(l)
    k = np.array(k)
    k *= pi0[None,:]
    k /= np.sum(k, axis = 1)[:,None]
    pi0 = np.average(k, axis = 0)
    mu0 = np.sum(y[:, None] * k, axis = 0) / (k.sum(axis = 0) * M)
    
print("Finished after %d iterations" % iter)
```
